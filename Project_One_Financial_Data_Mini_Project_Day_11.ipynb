```python
df.groupby('loan_status').agg({
    'id' : 'count',
    'dti' : 'mean',
    'int_rate' : 'mean',
    'loan_amount' : 'sum',
    'total_payment' : 'sum'
})
```

```python
loan_status_data = df.groupby('loan_status').agg({
    'id' : 'count',
    'dti' : 'mean',
    'int_rate' : 'mean',
    'loan_amount' : 'sum',
    'total_payment' : 'sum'
})
loan_status_data.rename(columns = {'id' : 'Number of Application'})
```

```python
loan_status_data = df.groupby('loan_status').agg({
    'id' : 'count',
    'dti' : 'mean',
    'int_rate' : 'mean',
    'loan_amount' : 'sum',
    'total_payment' : 'sum'
})
loan_status_data = loan_status_data.rename(columns={'id': 'No of Applications'})
good_loan_app = loan_status_data.loc['Fully Paid' , 'No of Applications'] + loan_status_data.loc['Current' , 'No of Applications']
print(f"Good Loan Applications: {good_loan_app}")
funded_amount = loan_status_data.loc['Fully Paid' , 'loan_amount'] + loan_status_data.loc['Current' , 'loan_amount']
print(f"Funded Amount: {funded_amount}")
receiver_amt = loan_status_data.loc['Fully Paid' , 'total_payment'] + loan_status_data.loc['Current' , 'total_payment']
print(f"Received Amount: {receiver_amt}")
print(f"Profit: {receiver_amt - funded_amount}")
```

```python
bad_loan_app = loan_status_data.loc['Charged Off' , 'No of Applications']
print(f"Bad Loan Applications: {bad_loan_app}")
bad_funded_amount = loan_status_data.loc['Charged Off' , 'loan_amount']
print(f"Bad Loan Funded Amount: {bad_funded_amount}")
bad_receiver_amt = loan_status_data.loc['Charged Off' , 'total_payment']
print(f"Bad Loan Received Amount: {bad_receiver_amt}")
bad_profit_loss = bad_receiver_amt - bad_funded_amount
print(f"Bad Loan Profit/Loss: {bad_profit_loss}")
```

```python
monthly_loan = df.groupby(df['issue_date'].dt.month)['loan_amount'].sum()
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.plot(months, monthly_loan.values)
plt.xlabel('Months')
plt.ylabel('Loan_amt')
```

```python
df["month"] = df["issue_date"].dt.month
new_df = df.groupby("month",as_index=False)["loan_amount"].sum()
new1_df = df.groupby("month",as_index=False)["total_payment"].sum()
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
x = months
y = new_df["loan_amount"]
plt.plot(x,y, marker = 'o')
plt.plot(x,new1_df["total_payment"], marker = 'o')
plt.title("Issue Date and Total Payment Analysis")
plt.xlabel("Issue Month")
plt.ylabel("Total Payment")
plt.grid()
plt.show()
```
