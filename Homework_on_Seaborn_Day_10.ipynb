# Using the Seaborn (sns) library, plot all the graphs.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_data.csv")
```

***Line Plot***

```python
sns.lineplot(data = df, x = 'study_hours', y = 'exam_score', color = 'gold', marker = 'o')
plt.title('Treand of Exam Scores by Study Hours of Students')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x='study_hours', y='exam_score', hue='placement_status', style='placement_status', markers=True)
plt.title("Exam Score Trends by Placement Status of Students")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

 ***Bar Plot***

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='placement_status', y='exam_score', palette='Set2')
plt.title("Average Exam Score by Placement")
plt.xlabel("Placement Status")
plt.ylabel("Exam Score (with Error Bars)")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='placement_status', y='exam_score', palette='mako', hue = 'study_hours')
plt.title("Average Exam Score by Placement")
plt.xlabel("Placement Status")
plt.ylabel("Exam Score (with Error Bars)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='sleep_hours', y='exam_score', hue='placement_status', palette='Set3')
plt.title("Exam Scores by Sleep Hours and Placement")
plt.xlabel("Hours of Sleep per Night")
plt.ylabel("Average Exam Score")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title="Placement Status")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='placement_status', palette='rocket', hue = 'study_hours')
plt.title("Total Number of Students by Placement Status")
plt.xlabel("Placement Status")
plt.ylabel("Total Count of Students")
plt.tight_layout()
plt.show()
```

***Scattter Plot***

```python
plt.figure(figsize=(9, 6))
sns.scatterplot(data=df, x='previous_score', y='exam_score', hue='placement_status', palette='YlOrBr', alpha=0.5)
plt.title("Previous Score vs. Exam Score by Placement")
plt.xlabel("Previous Score")
plt.ylabel("Final Exam Score")
plt.show()
```

***Histogram Plot***

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='exam_score', bins=20, color='orange', edgecolor='yellow')
plt.title("Distribution of Exam Scores")
plt.xlabel("Exam Score")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='exam_score', bins=20, kde=True, color='orange', edgecolor='gold')
plt.title("Distribution of Exam Scores")
plt.xlabel("Exam Score")
plt.ylabel("Frequency")
plt.show()
```

***Box plot***

```python
plt.figure(figsize=(9, 6))
sns.boxplot(data=df, x='placement_status', y='exam_score', palette='vlag', hue = 'study_hours')
plt.title("Exam Score Distributions by Placement Status")
plt.xlabel("Placement Status")
plt.ylabel("Exam Score")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x='study_hours', color='cyan')
plt.title("Distribution of Study Hours")
plt.xlabel("Study Hours per Day")
plt.show()
```

***Pie Chart Plot***

```python
placement_counts = df['placement_status'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(placement_counts, labels=placement_counts.index,
        autopct='%1.1f%%',
        colors=['gold', 'yellow'],
        explode=(0.05, 0),
        startangle=90,
        shadow=True)
plt.title("Proportion of Students Placed")
plt.show()
```

***Heatmap Plot***

```python
pivot_df = df.pivot_table(values='exam_score', index='sleep_hours', columns='study_hours', aggfunc='mean')
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_df, cmap='icefire', annot=True, fmt=".1f")
plt.title("Average Exam Score by Sleep & Study Hours")
plt.xlabel("Study Hours")
plt.ylabel("Sleep Hours")
plt.show()
```

```python
pivot_df = df.pivot_table(values='exam_score', index='sleep_hours', columns='study_hours', aggfunc='mean')
plt.figure(figsize=(18, 12))
sns.heatmap(pivot_df, cmap='flare', annot=True, fmt=".1f")
plt.title("Average Exam Score by Sleep & Study Hours")
plt.xlabel("Study Hours")
plt.ylabel("Sleep Hours")
plt.show()
```
