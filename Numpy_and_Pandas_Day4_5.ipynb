```python
import numpy as np
```

```python
a = np.random.randint(1,91, 25)
x = a.reshape(5,5)
mean_row = np.mean(x, axis = 1)
# finding the mean of the matrix row wise and the column wise
print(mean_row)
mean_column = np.mean(x, axis = 0)
print(mean_column)
#replacing all even elements with -1
replace = np.where(x % 2 == 0, -1, x)
print(replace)
# sort the array row wise and the coloumn wise
sort_row = np.sort(replace, axis = 1)
print(sort_row)
sort_column = np.sort(replace, axis = 0)
print(sort_column)
# find the sum of sorted row and sorted column
sorted_row_sum = np.sum(sort_row, axis = 1)
print(sorted_row_sum)
sorted_column_sum = np.sum(sort_column, axis = 0)
print(sorted_column_sum)
# for counting the number of -ve the each coloumns
neg_count = np.sum(replace < 0, axis = 0)
print(neg_count)
neg_col_count = np.sum(np.where(replace < 0, 1, 0), axis = 0)
print(neg_col_count)
```

```python
stocks = np.random.randint(20,80, size = (5,5))
print(stocks)
demand = np.random.randint(20,80, size = (5,5))
print(demand)
remaining_stock = stocks - demand
print(remaining_stock)
new_stock = np.where(remaining_stock < 0, 0, remaining_stock)
print(new_stock)
total_shortage = np.sum(np.where(remaining_stock < 0, remaining_stock * -1, 0), axis = 0)
print(total_shortage)
products_with_zero_stock = np.sum(np.where(total_shortage == 0, 1, 0), axis = 0)
print(products_with_zero_stock)
sort_based_on_remaining_stock = np.sort(remaining_stock, axis = 1)
print(sort_based_on_remaining_stock)
sort_based_on_remaining_stock = np.sort(remaining_stock, axis = 0)
print(sort_based_on_remaining_stock)
```
