```python
class complex:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def __add__(self, other):
    return complex(self.x + other.x, self.y + other.y)

  def __str__(self):
        return f"{self.x} + i{self.y}"

a = complex(2, 3)
b = complex(4, 5)

c = a + b
print(c.x, c.y)
```
