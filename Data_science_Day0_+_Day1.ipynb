Tuple Functions

count values - count()

find - index()

size - len()

total - sum()

highest/lowest - max()/min()

sort - sorted() [returns list]

```python
Tup = (20,43,95,76,81,20)
print(Tup.count(20)) #count - used to tell how many times a number returns
print(Tup.index(95)) #index - used to find the index of a value provided
print(len(Tup)) #len - tells the length of the tuple
print(sum(Tup)) #sum - sum of all the elements in the tuple
print(max(Tup)) #max - returns the maximum element
print(min(Tup)) #min - returns the maximum element
print(sorted(Tup)) #sorted - returns list after sorting a tuple
```
