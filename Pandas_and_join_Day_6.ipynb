#1757 question - Leetcode



```python
import pandas as pd
```

```python
dic = {
    "product_id" : [100,101,102,103],
    "low_fats" : ["Y","Y","N","Y"],
    "recyclable" : ["N","Y","Y","Y"]
}
df = pd.DataFrame(dic)
display(df)
Answer = (df['low_fats'] == 'Y') & (df['recyclable'] == 'Y')
print("Boolean Series 'Answer':")
print(Answer)
filtered_product_ids = df[Answer]['product_id']
print("\nProduct IDs where both conditions are true:")
print(filtered_product_ids)
```

#584 question - Leetcode


```python
dic = {
    "id" : [1,2,3,4,5,6],
    "name" : ["Will","Jane","Alex","Bill","Zack","Mark"],
    "referee_id" : [None , None , 2 , None , 1 , 2]
}
df = pd.DataFrame(dic)
display(df)
result_df = df[(df['referee_id'].isnull()) | (df['referee_id'] != 2)]
final_names = result_df[['name']]
display(final_names)
```

#Continuing with pandas

```python
df = pd.read_csv('student.csv')
```

```python
df.groupby('placement_status')['placement_status'].agg('count')
```

```python
df.groupby('placement_status')['exam_score'].agg('mean')
```

```python
df.groupby('placement_status')['exam_score'].mean()
```

```python
df.groupby('exam_score')['study_hours'].sum()
```

```python
def grades(score):
  if score >= 90:
    return 'A'
  elif score >= 80:
    return 'B'
  elif score >= 70:
    return 'C'
  elif score >= 60:
    return 'D'
  else:
    return 'E'

df['grade'] = df['exam_score'].apply(grades)
df
```
